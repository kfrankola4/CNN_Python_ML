```python
#loading all the pickles

#trainingP = pkl.load(open('training_P.pkl', 'rb'))
#trainingS = pkl.load(open('training_S.pkl', 'rb'))

#now we have a single pickle tying all of the training data together
training = pkl.load(open('training2.pkl', 'rb'))
testing = pkl.load(open('testing.pkl', 'rb'))



```

```python
#now I'm going to read the dictionary which in essence will reconstruct the
#pickles (which are dictionaries) based on how I packaged them

#need to iterate through all elements in each pickle
#trainingP_reconstructed = []
#trainingS_reconstructed = []

#again now that there is only one training pickle we need to reorganize the function to unpack it

testing_reconstructed = []
training_reconstructed = []


#ignore::::
#_________________________________________
#for element in trainingP:
#    trainingP_reconstructed.append(Image.frombytes(element['mode'], element['size'], element['pixels']))

#for element in trainingS:
#    trainingS_reconstructed.append(Image.frombytes(element['mode'], element['size'], element['pixels']))
#_________________________________________


#i want to iterate over the list then iterate over each sub list extracting the image from each ELEMENT of 
#of the sub list so that it is extracted then put back into the training_reconstructed list

training_sub_iterator = 0
for element in training:
    for element in training[training_sub_iterator]:
        training_sub_list_pair = []
        reconstruced = Image.frombytes(element['mode'], element['size'], element['pixels'])
        training_sub_list_pair.append(reconstruced)
    training_reconstructed.append(training_sub_list_pair)
    training_sub_iterator = training_sub_iterator+1
        
    
for element in testing:
    testing_reconstructed_sub = Image.frombytes(element['mode'], element['size'], element['pixels'])
    testing_reconstructed.append(testing_reconstructed_sub)


```
